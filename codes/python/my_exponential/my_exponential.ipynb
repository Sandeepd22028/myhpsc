```python
import numpy as np
import matplotlib.pyplot as plt
import math

tol=1.0e-14

def my_exp(k,n=100,debug=False):
    y=0
    for i in range(0,n):
        temp=y
        y=y+k**i/np.math.factorial(i)
        if debug:
            if(abs(y-temp)<tol):
                print(f"series is converged at i={i}")
                break
    return y
```

```python
x=10
my_exp(x,100,True)
```

```python
x=np.linspace(0,100,1000)
y=my_exp(x)
z=np.exp(x)
plt.plot(x,y,'-r',label="my_exp(x)")
plt.plot(x,z,'-g',label="numpy_exp(x)")
plt.legend()
plt.show()
```

```python
def k(x):
    y=my_exp(x)
    return y
def l(v):
    s=np.exp(v)
    return s
%timeit k(10)
%timeit l(10)
```
